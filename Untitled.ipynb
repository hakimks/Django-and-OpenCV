```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('mydog.jpg')
img
```

```python
type(img)
```

```python
img.shape
```

```python
plt.imshow(img)
```

```python
normal = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(normal)
```

```python
def get_image(image, cmap=None):
    plt.axis('off')
    plt.imshow(image, cmap=cmap)
    plt.show()
```

```python
get_image(normal)
```

## Colored

```python
hav = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
get_image(hav)
```

## Greyscale

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
get_image(gray, cmap='gray')
```

## Blured

```python
k4 = (25,25)
blur4 = cv2.blur(normal, k4)
get_image(blur4)
```

```python
k2 = (2,2)
blur2 = cv2.blur(normal, k2)
get_image(blur2)
```

```python
k3 = (100,100)
blur3 = cv2.blur(normal, k3)
get_image(blur3)
```

#### need to apply kernel by image size

```python
img.shape
```

```python
width,height = img.shape[:2]
print(width)
print(height)

if width > 500:
    kn = (50, 50)
elif width > 200 and width <= 500:
    kn = (25,25)
else:
    kn = (10, 10)

blur = cv2.blur(normal, kn)
get_image(blur)
```

# Binary

```python
v, binary = cv2.threshold(gray, 100, 255, cv2.THRESH_BINARY)
get_image(binary, cmap='gray')
```

## Invert

```python
invert = cv2.bitwise_not(binary)
get_image(invert, cmap='gray')
```
